```python
class Individuo:
    def __init__(self,genes):
        self.genes = genes
        self.fitness = 0

    def evaluar(self):
        self.fitness = sum(self.genes)

    def mutar(self):
        #Cambia el 1er gen (0 a 1 o viceversa)
        if self.genes[0] == 0:
            self.genes[0] = 1
        else:
            self.genes[0] = 0
```

```python
class Poblacion:
    def __init__(self,individuos):
        self.individuos = individuos

    def evaluar(self):
        for individuo in self.individuos:
            individuo.evaluar()

    def seleccionar_mejor(self):
        mejor = self.individuos[0]
        for individuo in self.individuos:
            if individuo.fitness > mejor.fitness:
                mejor = individuo
        return mejor

    def cruzar(self,papa,mama):
        #Cruce simple: mitad y mitad
        mitad = len(papa.genes) // 2
        nuevos_genes = papa.genes[:mitad] + mama.genes[mitad:]
        return Individuo(nuevos_genes)
```

```python
# Individuos iniciales
i1 = Individuo([1, 0, 1])
i2 = Individuo([0, 1, 0])
i3 = Individuo([1, 1, 0])

# Población
p = Poblacion([i1, i2, i3])

# Evaluar
p.evaluar()

# Seleccionar mejor
mejor = p.seleccionar_mejor()
print("Mejor:", mejor.genes, mejor.fitness)

# Cruzar dos individuos
hijo = p.cruzar(i1, i2)
hijo.evaluar()
print("Hijo:", hijo.genes, hijo.fitness)

# Mutar hijo
hijo.mutar()
hijo.evaluar()
print("Hijo mutado:", hijo.genes, hijo.fitness)
```
