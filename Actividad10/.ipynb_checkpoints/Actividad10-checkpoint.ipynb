# Preguntas:
1. El optimizador de algoritmo utilizado es Adam, pero existen otros como SGD y RMSProp, describe cada uno de ello y explica como pudieran cambiar los resultados de tu modelo.
- SGD (Descenso de Gradiente EstocГЎstico): actualiza los pesos con cada dato individualmente. Es simple pero lento y puede quedarse atascado en mГ­nimos locales.
- RMSProp: adapta el learning rate por cada parГЎmetro. Funciona mejor que SGD en problemas con datos variables.
- Adam (el usado): combina lo mejor de ambos. Es el mГЎs estable y rГЎpido de los tres.
Con SGD el modelo probablemente tardarГ­a mГЎs epochs en converger o no llegarГ­a al mismo nivel de precisiГіn.

2. Describe porque es importante agregar una funciГіn de pГ©rdida como 'mean_squared_error ', investiga que otra funciГіn de pГ©rdida se puede implementar y como alterarГ­a los resultados.
- mean_squared_error es importante porque penaliza fuertemente los errores grandes, guiando al modelo hacia predicciones mГЎs precisas.
Otra opciГіn que encontre es mean_absolute_error, que mide el error promedio sin elevar al cuadrado. Es menos sensible a valores atГ­picos, pero el modelo converge mГЎs lento y con menor precisiГіn en este caso.

3. En las modificaciones realizadas,  cuantas epocas fueron 'suficientes' para que el modelo se considere entrenado para ponerlo en funcionamiento.

4. ВїPor quГ© son importantes las epoch?
- Porque cada epoch permite al modelo ajustar sus pesos y reducir el error. Pocas epochs es un modelo sin entrenar, pero demasiadas es un riesgo de overfitting.

- El modelo actualmente no tiene funciГіn de activaciГіn definida, lo que significa que usa lineal por defecto, que es adecuada para problemas como este.
