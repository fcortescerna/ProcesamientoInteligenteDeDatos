```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#importar datos
df = pd.read_csv("CincinnatiPaintCompany.csv")
df.head()
```

### 1. Determinar las variables independientes
x1 = calls, x2 = driven

```python
X = df[['Calls','Driven']]
```

### 2. Determinar las variables dependientes Y
y = commissions

```python
Y = df[['Commissions']]
```

```python
#Ajustar el model de regresión multiple
model = LinearRegression()
model.fit(X, Y) #Entrena el modelo
y_pred = model.predict(X)
#y_pred

#Métricas estadísticas
mse = mean_squared_error(Y, y_pred)
r2 = r2_score(Y, y_pred)
```

### 3. Valor de intersección

```python
model.intercept_
```

```python
#coeficientes
print('Coeficientes: ', model.coef_)
print("Error cuadrático medio ", mse)
print("Varianza: ", r2)
```
