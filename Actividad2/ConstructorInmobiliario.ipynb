# Regresión Líneal Múltiple - Constructor Inmobiliario

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#importar datos
df = pd.read_csv('ConstructorInmobiliario.csv')
df.head()
```

# 1. Determinar variables independientes
En el primer modelo, no se incluye si la columna si había un adulto mayor viviendo en la casa, dado a que en el tamaño de la familia viene incluido. No se incluye la diferencia de educación entre la pareja, dado que queremos probar si no es relevante para la elección de la casa

```python
X1 = df[['Income', 'Size']]
X2 = df[['Income', 'Size', 'Senior', 'Education']]
df[['Income', 'Size', 'Senior', 'Education']]
```

# 2. Determinar variables dependientes

```python
Y = df[['SqrFt']]
df[['SqrFt']]
```

# 4. Modelo 1 Simple (Income and Size)

```python
#Ajustar el primer modelo de regresión múltiple
model1 = LinearRegression() 
model1.fit(X1, Y) #Entrenando el modelo 
y1_pred = model1.predict(X1)
y1_pred
```

# 5. Modelo 2 Completo (Income, Size, Senior, Education)

```python
#Ajustar el primer modelo de regresión múltiple
model2 = LinearRegression() 
model2.fit(X2, Y) #Entrenando el modelo 
y2_pred = model2.predict(X2)
y2_pred
```

# 6. Métricas Simple (Income and Size)

```python
#Metricas estadísticas
mse1 = mean_squared_error(Y, y1_pred) #Error de la media cuadrática
r21 = r2_score(Y, y1_pred) #varianza
```

```python
print('Valor de intersección : ', model1.intercept_)
print('Coeficientes: ', model1.coef_)
print('Error cuadrático medio', mse1)
print('Varianza: ', r21)
```

# 7. Métricas Completo (Income, Size, Senior, Education)

```python
#Metricas estadísticas
mse2 = mean_squared_error(Y, y2_pred) #Error de la media cuadrática
r22 = r2_score(Y, y2_pred) #varianza
```

```python
print('Valor de intersección : ', model2.intercept_)
print('Coeficientes: ', model2.coef_)
print('Error cuadrático medio', mse2)
print('Varianza: ', r22)
```

# 8. Conlusiones
Aunque el primer modelo presenta un buen ajuste y simplicidad, el segundo modelo demuestra un mejor rendimiento predictivo y una mayor capacidad para explicar la variabilidad de los datos.
